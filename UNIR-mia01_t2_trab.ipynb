***1.	Crea una función en Python que calcule el máximo común divisor de dos números.***

> _El MCD es el máximo número entero que divide un conjunto de dos o más números enteros sin dejar residuo._

Definimos la función mcd

```python
# define la función mcd
def mcd(a, b):
    resto = 0
    while(b > 0):
        resto = b
        b = a % b
        a = resto
    return a
```

Solicitamos los 2 valores y calcualmos el resultado.

```python
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
print("El máximo común divisor de ", num1," y ", num2," es ", mcd(num1, num2))
```

**2.	Crea una función en Python que calcule el mínimo común múltiplo de dos números.**
> _El MCM Es el menor número natural distinto de cero que es múltiplo común de un conjunto de 2 o más números naturales_

```python
# define la función gcd (recursiva) y mcm
def gcd(x, y):
    return x if not y else gcd(y, x % y)
 
def mcm(x, y):
    return x * y / gcd(x, y)
```

Solicitamos los 2 valores y calcualmos el resultado.

```python
num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))
print("El minimo comun multiplo de ", num1," y ", num2," es ", mcm(num1, num2))
```

### Segunda parte: Procesado y limpieza de datos con funciones de cadena ###

Primero definiremos una función para eliminar tildes. No afecta a las ñ, como lo hacen algunas funciones con unicodedata

```python
def elimina_tildes(cadena):
    a,b = 'áéíóúü','aeiouu'
    trans = str.maketrans(a,b)
    return cadena.translate(trans)    
    return cadena
```

**Ahora la actividad. Dentro de la siguiente porción de código incluiremos 4 tareas, que no podemos separar dado que dependen de la lectura linea a linea del fichero.**

1.	Lee el fichero proporcionado.
2.	Cuenta el número de líneas del fichero (contador_linea).
3.	Cuenta el número de veces que aparece el artículo «el» en el fichero (nota: elemento contiene los caracteres «el» pero no debe contar puesto que no contiene a el artículo «el».
4.	Elimina las tildes del fichero (elimina_tildes) y convierta todas las letras a minúscula (linea_lower)

```python
# abre fichero texto, modalidad lectura
f = open("texto.txt", "r")

# define variables de trabajo. COntador de llineas y un array "nueva_linea" para escribir en el nuevo texto.
contador_linea = 0 # inicializa en 0
ocurrencias = 0 # inicializa en 0
nueva_linea = []

# lee el texto linea a linea 
for linea in f:
        
    #contador de lineas del texto
    contador_linea = contador_linea + 1
           
    #conversor linea a minusculas. 
    linea_lower = linea.lower()
    
    #elimina tildes
    linea_sin_tildes = elimina_tildes(linea_lower)
    
    #print(linea_sin_tildes)
    
    nueva_linea.append(linea_sin_tildes)
    
    #COntador de cocurrencias de articulo "el" ya en minusculas y sin tildes
    ocurrencias = ocurrencias + linea.count("el ")
    
    
f.close()
```

```python
# abre un fichero de escritura
f = open('resultado.txt', 'w')

try:
    # escribe en las 2 primeras lilneas del texto:
    f.write("Numero de lineas del fichero text.txt                     : " + str(contador_linea) + "\n" )
    f.write("Numero de apariciones de alticulo -el- en fichero text.txt: " + str(ocurrencias) + "\n") 
        
    # Procesamiento para escribir en el fichero linea a linea el array con lineas minusculas y sin tildes
    for nl in range(len(nueva_linea)):
        f.write(nueva_linea[nl])
finally:
    f.close()
```
